```python
import ast
```

```python
def convert(obj):
    L= []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert1(obj):
    L= []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies
```

```python
def fetch_director(obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
           L.append(i['name'])
           break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

## BAG OF WORDS


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]  #getting movie index
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) #for each indexed getting distance
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=8000,
    stop_words='english',
    ngram_range=(1,2),
    min_df=2,
    max_df=0.8
)

vectors_new = tfidf.fit_transform(new_df['tags']).toarray()
similarity_new = cosine_similarity(vectors_new)
```

```python
similarity_new
```

```python
def recommend_new(movie, top_n=5):
    movie = movie.lower()
    
    if movie not in new_df['title'].str.lower().values:
        return "Movie not found"
    
    index = new_df[new_df['title'].str.lower() == movie].index[0]
    distances = sorted(
        list(enumerate(similarity_new[index])),
        reverse=True,
        key=lambda x: x[1]
    )
    
    recommendations = []
    for i in distances[1:top_n+1]:
        recommendations.append(new_df.iloc[i[0]].title)
        
    return recommendations
```

```python
recommend_new('Avatar')
```

```python
recommend('Batman Begins')   # using BoW
```

```python
recommend_new("Batman Begins")   # using TF-IDF
```
